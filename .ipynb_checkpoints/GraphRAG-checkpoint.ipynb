````python
def solve_extracted_json(extracted_json):
    # N·∫øu d·ªØ li·ªáu l√† list, l·∫•y ph·∫ßn t·ª≠ ƒë·∫ßu ti√™n
    if isinstance(extracted_json, list) and len(extracted_json) > 0:
        extracted_json = extracted_json[0]

    # Ki·ªÉm tra n·∫øu v·∫´n kh√¥ng ph·∫£i l√† string, c·∫ßn x·ª≠ l√Ω
    if not isinstance(extracted_json, str):
        raise ValueError("extract_entities() ph·∫£i tr·∫£ v·ªÅ m·ªôt chu·ªói JSON h·ª£p l·ªá.")

    # Lo·∫°i b·ªè Markdown ```json ... ```
    cleaned_json = re.sub(r'```json\n|\n```', '', extracted_json)

    # Chuy·ªÉn ƒë·ªïi th√†nh dictionary
    data = json.loads(cleaned_json)

    # Chu·∫©n h√≥a d·ªØ li·ªáu
    def normalize_value(value):
        if isinstance(value, str):
            if value.lower() == "not mentioned":
                return None
            if value.isdigit():
                return int(value)
            if ">" in value:
                return {"operator": ">", "value": int(value.replace(">", ""))}
        return value

    data = {key: normalize_value(value) for key, value in data.items()}

    return data

# G·ªçi h√†m
extracted_json = extract_entities(user_input)  # H√†m n√†y c·∫ßn ƒë·∫£m b·∫£o tr·∫£ v·ªÅ m·ªôt chu·ªói ho·∫∑c danh s√°ch ch·ª©a chu·ªói JSON

cleaned_data = solve_extracted_json(extracted_json)
print(json.dumps(cleaned_data, indent=2))
````

````python
def generate_cypher_query(json_data):
    query_base = """
    MATCH (h:House)-[r]->(e)
    WHERE 
    """
    
    conditions = []
    params = {}
    
    # ƒê·ªãnh nghƒ©a c√°c m·ªëi quan h·ªá
    relationships = {
        "Location": "LOCATED_IN",
        "Ownership": "HAS_OWNERSHIP",
        "Status": "HAS_STATUS",
        "Transaction": "HAS_TRANSACTION",
        "Furnishing": "HAS_FURNISHING",
        "Facing": "HAS_FACING",
        "Bathroom": "HAS_BATHROOM",
        "Balcony": "HAS_BALCONY",
        "CarParking": "HAS_CAR_PARKING",
        "CurrentFloor": "HAS_CURRENT_FLOOR",
        "TotalFloors": "HAS_TOTAL_FLOORS"
    }
    
    for key, value in json_data.items():
        if value is None:
            continue  # B·ªè qua gi√° tr·ªã null
        
        relationship = relationships.get(key)
        if not relationship:
            continue  # B·ªè qua n·∫øu kh√¥ng t√¨m th·∫•y quan h·ªá
        
        if isinstance(value, dict) and "operator" in value and "value" in value:
            # X·ª≠ l√Ω ƒëi·ªÅu ki·ªán v·ªõi to√°n t·ª≠ (>, <, >=, <=, =)
            operator = value["operator"]
            param_key = f"{key}_val"
            conditions.append(f"(h)-[:{relationship}]->(e) AND e.value {operator} ${param_key}")
            params[param_key] = value["value"]
        else:
            # X·ª≠ l√Ω ƒëi·ªÅu ki·ªán th√¥ng th∆∞·ªùng
            param_key = f"{key}_val"
            conditions.append(f"(h)-[:{relationship}]->(e) AND e.name = ${param_key}")
            params[param_key] = value
    
    query_conditions = "\n    AND ".join(conditions)
    query_full = query_base + query_conditions + """\n    RETURN 
        h.title AS house_title, h.description AS house_description, 
        h.price AS house_price, h.amount AS house_amount, 
        type(r) AS relationship, 
        labels(e)[0] AS entity_type, e.name AS entity_name 
    LIMIT 100"""
    
    return query_full, params

# JSON ƒë·∫ßu v√†o
json_input = {
    "Title": "1 BHK Ready to Occupy Flat for sale in Virat",
    "Location": "Thane",
    "Status": "Ready to Move",
    "Transaction": "Resale",
    "Furnishing": "Unfurnished",
    "Facing": None,
    "Bathroom": 2,
    "Balcony": None,
    "CarParking": "1 open",
    "Ownership": None,
    "CurrentFloor": {"operator": ">", "value": 0},
    "TotalFloors": {"operator": ">", "value": 10}
}

# T·∫°o truy v·∫•n
query, params = generate_cypher_query(json_input)
print("Cypher Query:")
print(query)
print("\nQuery Parameters:")
print(params)
````

````python
# Sinh truy v·∫•n t·ª´ JSON ƒë·∫ßu v√†o
query, params = generate_cypher_query(json_input)

# Th·ª±c thi truy v·∫•n tr√™n Neo4j
result = graph.query(query, params)

# X·ª≠ l√Ω d·ªØ li·ªáu tr·∫£ v·ªÅ
house_entities = {}
for record in result:
    house_title = record["house_title"]
    if house_title not in house_entities:
        house_entities[house_title] = {
            "description": record["house_description"],
            "price": record["house_price"],
            "amount": record["house_amount"],
            "entities": []  # Danh s√°ch th·ª±c th·ªÉ li√™n k·∫øt
        }
    house_entities[house_title]["entities"].append({
        "type": record["entity_type"],
        "name": record["entity_name"],
        "relationship": record["relationship"]
    })

# In k·∫øt qu·∫£
import json
print(json.dumps(house_entities, indent=2, ensure_ascii=False))
````

````python
MATCH (h:House)-[:LOCATED_IN]->(loc:Location)
MATCH (h)-[:HAS_STATUS]->(status:Status)
MATCH (h)-[:HAS_TRANSACTION]->(transaction:Transaction)
MATCH (h)-[:HAS_FURNISHING]->(furnishing:Furnishing)
MATCH (h)-[:HAS_CAR_PARKING]->(parking:CarParking)
MATCH (h)-[:HAS_BATHROOM]->(bathroom:Bathroom)
//MATCH (h)-[:HAS_CURRENT_FLOOR]->(floor:Floor)
// MATCH (h)-[:HAS_TOTAL_FLOORS]->(totalFloors:TotalFloors)
WHERE 
  loc.name = 'thane'
  AND status.name = 'Ready to Move'
  AND transaction.name = 'Resale'
  AND furnishing.name = 'Unfurnished'
  AND parking.name = '1 Open'
  AND bathroom.name ="1"
 // AND floor.name > 0
  // AND totalFloors.name > 10
RETURN 
  h.title AS house_title, h.description AS house_description, 
  h.price AS house_price, h.amount AS house_amount, 
  labels(loc)[0] AS location_type, loc.name AS location_name, 
  labels(status)[0] AS status_type, status.name AS status_name, 
  labels(transaction)[0] AS transaction_type, transaction.name AS transaction_name,
  labels(furnishing)[0] AS furnishing_type, furnishing.name AS furnishing_name,
  labels(parking)[0] AS parking_type, parking.name AS parking_name,
  labels(bathroom)[0] AS bathroom_type, bathroom.name AS bathroom_number 
  //labels(floor)[0] AS floor_type, floor.name AS floor_name
  // labels(totalFloors)[0] AS totalFloors_type, totalFloors.name AS totalFloors_number 
LIMIT 100
````
