```python
import pandas as pd 
```

```python
df = pd.read_csv("data/house_prices.csv")
df.head(5)
```

```python
# Tính số lượng NaN và phần trăm NaN của từng cột
nan_counts = df.isnull().sum()  # Số lượng NaN
nan_percentage = (nan_counts / len(df)) * 100  # Phần trăm NaN

# Tạo DataFrame để dễ đọc kết quả
nan_summary = pd.DataFrame({"Missing Values": nan_counts, "Percentage (%)": nan_percentage})

# Hiển thị kết quả
print(nan_summary.sort_values(by="Percentage (%)", ascending=False))
```

```python
df["facing"] = df["facing"].fillna("Unknown")
df["overlooking"] = df["overlooking"].fillna("Unknown")
df["Balcony"] = df["Balcony"].fillna(0)
df["Ownership"] = df["Ownership"].fillna("Unknown")
df["Furnishing"] = df["Furnishing"].fillna("Unknown")
df["Description"] = df["Description"].fillna("Don't have description")
df["Car Parking"] = df["Car Parking"].fillna(0)
df["Status"].fillna(df["Status"].mode()[0], inplace=True)
df["Floor"].fillna(df["Floor"].mode()[0], inplace=True)
df["Bathroom"] = df["Bathroom"].fillna(1)
df.drop(columns=["Plot Area"], inplace=True)
df.drop(columns=["Dimensions"], inplace=True)
df.drop(columns=["Society"], inplace=True)
df.drop(columns=["Super Area"], inplace=True)
df.drop(columns=["overlooking"], inplace=True)
df.drop(columns=["Carpet Area"], inplace=True)

```

```python
# Tính phần trăm từng giá trị trong cột "Status"
status_counts = df["Transaction"].value_counts(normalize=True) * 100

# In kết quả
print(status_counts)
```

```python
df.dropna(subset=["Price (in rupees)"], inplace=True)
df.dropna(subset=["Transaction"], inplace=True)
```

```python
# Tính số lượng NaN và phần trăm NaN của từng cột
nan_counts = df.isnull().sum()  # Số lượng NaN
nan_percentage = (nan_counts / len(df)) * 100  # Phần trăm NaN

# Tạo DataFrame để dễ đọc kết quả
nan_summary = pd.DataFrame({"Missing Values": nan_counts, "Percentage (%)": nan_percentage})

# Hiển thị kết quả
print(nan_summary.sort_values(by="Percentage (%)", ascending=False))
```

```python
df[["Current Floor", "Total Floors"]] = df["Floor"].str.extract(r"(\d+) out of (\d+)")
df["Current Floor"] = pd.to_numeric(df["Current Floor"], errors="coerce").fillna(0).astype(int)
df["Total Floors"] = pd.to_numeric(df["Total Floors"], errors="coerce").fillna(0).astype(int)
df.drop(columns=["Floor"], inplace=True)
```

```python
# Hàm chuyển đổi từ "Lac" hoặc "Cr" sang rupee
def convert_to_rupees(value):
    if pd.isna(value):  # Xử lý NaN
        return None
    value = str(value).replace(",", "").strip()  # Xóa dấu phẩy nếu có
    if "Lac" in value:
        return float(value.replace("Lac", "").strip()) * 1_00_000  # Nhân với 100,000
    elif "Cr" in value:
        return float(value.replace("Cr", "").strip()) * 1_00_00_000  # Nhân với 10,000,000
    
df["Amount(in rupees)"] = df["Amount(in rupees)"].apply(convert_to_rupees)
print(df[["Amount(in rupees)", "Price (in rupees)"]].head())
```

```python
# Tính số lượng NaN và phần trăm NaN của từng cột
nan_counts = df.isnull().sum()  # Số lượng NaN
nan_percentage = (nan_counts / len(df)) * 100  # Phần trăm NaN

# Tạo DataFrame để dễ đọc kết quả
nan_summary = pd.DataFrame({"Missing Values": nan_counts, "Percentage (%)": nan_percentage})

# Hiển thị kết quả
print(nan_summary.sort_values(by="Percentage (%)", ascending=False))
```

```python
df.dropna(subset=["Total Floors"], inplace=True)
df.dropna(subset=["Current Floor"], inplace=True)
```

```python
# Lưu lại dữ liệu đã xử lý
df.to_csv("data/cleaned_house_prices_data.csv", index=False)
```
